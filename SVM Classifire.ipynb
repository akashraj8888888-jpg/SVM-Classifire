```python
# Importing phase
```

```python
#importing the pandas libraries
import pandas as pd
```

```python
#read csv file for read pandas
df = pd.read_csv("Social_Network_Ads.csv")
```

```python
#print the variable for output
df
```

# One-Hot Encoding With `pd.get_dummies()`

```python
# one-hot encode categorical columns & drop first to avoid dummy trap
df = pd.get_dummies(df,dtype=int,drop_first=True)
```

```python
df
```

```python
# Axis =  1 --> Column wise delete
# Axis = 0 ---> Row wise delete
df = df.drop("User ID",axis = 1)
```

```python
 df
```

# Spliting process

```python
Independent = df[['Gender_Male','Age','EstimatedSalary']]
```

```python
Independent
```

```python
Dependent = df[['Purchased']]
```

```python
Dependent
```

# Train Test Split

```python
# splitting the dataset into training and testing
from sklearn.model_selection import train_test_split
```

```python
train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
# it is using for variables
x_train,x_test,y_train,y_test=train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
x_train,x_test,y_train,y_test
```

# Import the algorithm (train phase)

```python
from sklearn.svm import SVC
```

```python
Classifier = SVC(kernel='rbf',random_state=0)
```

```python
Classifier
```

```python
Classifier.fit(x_train, y_train)
```

```python
y_pred = Classifier.predict(x_test)
```

# To find the classification_report

```python
from sklearn.metrics import classification_report
```

```python
Clf_report = classification_report(y_test, y_pred)
```

```python
Clf_report
```

# Test the model using values

```python
Gender = int(input("Enter you gender (Male: 1, Female: 0): "))
Age = int(input("Enter you Age: "))
Estimated_Salary = int(input("Enter your Estimated Salary: "))

Classifier.predict([[Gender, Age, Estimated_Salary]])
```
